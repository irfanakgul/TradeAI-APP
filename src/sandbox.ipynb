```python
from context_pull import fn_read_from_db,fn_write_to_db
```

```python
df_raw_all = fn_read_from_db('raw_ticker_daily')
```

```python
usa_first2000 = (
    df_raw_all[df_raw_all["COUNTRY"] == "Turkey"]
      .groupby("TICKER", as_index=False)["VOLUME"]
      .sum()
      .rename(columns={"VOLUME": "sum_vol"})
      .sort_values("sum_vol", ascending=False)
      .head(500)
)
usa_first2000['COUNTRY'] = "Turkey"
usa_first2000['FILTER'] = "first500_vol"
```

```python
fn_write_to_db(usa_first2000,'turkey_first500_company','replace')
```
